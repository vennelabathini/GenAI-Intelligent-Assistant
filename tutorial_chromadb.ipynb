```python
import chromadb

client = chromadb.Client()
collection = client.create_collection(name="my_collection")

```

```python
collection.add(
        documents=["This document is about Python.", "This document is about AI."],

    ids=["id1", "id2"]
)
```

```python
all_docs = collection.get()
print(all_docs)
```

```python
collection.delete(ids=all_docs['ids'])
collection.get()
```

```python
collection.add(
        documents=["This document is about Python.","This document is about AI."],

    ids=["id3","id4"],
    metadatas=[{"url":"http://en.wikipedia.org/wiki/Artifical_intelligence"},{"url":"http://en.wikipedia.org/wiki/python_(programming_language)"}]
)
```

```python
results= collection.query(
    query_texts=["Tell me about AI."],
    n_results=1,
    include=["metadatas"]
)
print(results)
```
